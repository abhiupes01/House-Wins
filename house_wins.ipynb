```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create virtual die simulation, if the value is between 1-51 the house wins
# while if the value is between 52 - 100 the player wins
```

```python
# probability of player winning the game
prob_player = (100-51)/100
print('Probability of player winning :{} %'.format(prob_player*100))

# probability of house winning
prob_house = 1-prob_player
print('Probability of house winning :{} %'.format(prob_house*100))

# Expected profit for player
exp_prob = 1*prob_player-(1*prob_house)
print('Expected profit for player :{:4.2f} %'.format(exp_prob*100))
```

```python
# here the house is always on edge of winning the game
```

```python
# create function to simuate the die rollling

def rolling():
    roll = np.random.randint(1,100)
    if roll <= 51:
        return False
    elif roll >51 and roll <=100:
        return True
```

```python
# define a betting function, the func takes 3 arguments such as :
# 1) total amount = total money with which player starts betting
# 2) bet amount = betting amount player wages in each bet
# 3) total number of bets 

def betting(total_amount,bet_amount,total_bets):
    
    # create an empty list to record the number of bets and bet won and lost
    fund = []
    bets = []
    
    # start with bet 1
    bet = 1
    while bet <= total_bets:
        if rolling():
            total_amount += bet_amount
            fund.append(total_amount)
            bets.append(bet)
            
        else:
            total_amount -= bet_amount
            fund.append(total_amount)
            bets.append(bet)
            
        bet = bet +1
        
    # plot the values
    plt.plot(bets,fund)
    return fund[-1]
```

```python
def monte_carlo(num_times,total_amount,bet_amount,bets):
    

    i =1
    final_fund = []
    while i<=num_times:
        last_amount = betting(total_amount,bet_amount,bets)
        final_fund.append(last_amount)
        i = i+1
        
    plt.xlabel('Number of bets')
    plt.ylabel('Money with player')
    plt.title('Player started with {} and ended with {}'.format(10000,np.mean(final_fund)))
```

```python
monte_carlo(100,10000,100,10)
```
